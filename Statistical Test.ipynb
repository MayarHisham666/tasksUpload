Statistical Test:

Definition:
Statistic test is data analysis process that describes the relation between samples of data have significant hypothesis relation (difference predictor output variables) 
from null hypothesis data (no difference or relation between samples) which calculated by p-value (probability value).

Some of condations to validate the Statistical Test:


1-validation Data collection:

Dependens on:

i-Data collection process/source: (experiment, observation using probability sampling).

ii-Sample size needs to be large enough to approximate the true distribution of the population being studied.

iii-Meets certain assumptions.

2-Types of parametric test:

Regression tests:

Can be used to estimate the effect of one or more continuous variables on another variable.
Comparison:

Can be used to test the effect of a categorical variable on the mean value of some other characteristic.

As T-test,ANOVA and MANOVA tests

Correlation:

Estimate relation between data features.


3- Types of nonparametric test:

Are useful when one or more of the common statistical assumptions are violated.
