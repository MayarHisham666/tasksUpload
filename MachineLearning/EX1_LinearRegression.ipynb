```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices1.csv')
df
```

```python
print(df.area.min())
print(df.area.max())
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area,df.price,color='r',marker='x')
```

```python
area=df.drop('price',axis='columns')
area           
```

```python
price=df.price
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(area,price)
```

```python
reg.predict([[3000],[3750]])
```

```python
reg.coef_ # m coeff
```

```python
reg.intercept_ # y intercept
```

```python
area2= pd.read_csv('areas.csv')
area2
```

```python
p=reg.predict(area2)
p
```

```python
area2['price']=p
area2
```

```python
area2.to_csv("Prediction.csv",index=False)
```
